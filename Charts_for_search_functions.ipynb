```python
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt2

x  = [100000000, 200000000, 300000000, 400000000, 500000000]
y1 = [38.14, 69.75, 113.34, 143.46, 176.42]
plt.plot(x, y1, label="Search func")

plt.xlabel("size mil")
plt.ylabel("time milsec")
plt.title("BigO Search")
plt.grid(1)
plt.legend()
plt.show()

x2 = [2000000000, 2100000000, 2200000000, 2300000000, 2400000000]
y2 = [0.003, 0.004, 0.005, 0.009, 0.012]
plt2.plot(x2, y2, label="Binary search")

plt2.xlabel("size bil")
plt2.ylabel("time milsec")
plt2.title("BigO Binary search")
plt2.grid(1)
plt2.legend()
plt2.show()
```
