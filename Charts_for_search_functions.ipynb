```python
import matplotlib.pyplot as plt

x  = [100000000, 200000000, 300000000, 400000000, 500000000]
y1 = [38.14, 69.75, 113.34, 143.46, 176.42]
y2 = [0, 0, 0, 0, 0]
plt.plot(x, y1, label="Search func")
plt.plot(x, y2, label="Bin search undone")
plt.plot()

plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Line Graph Example")
plt.legend()
plt.show()
```
