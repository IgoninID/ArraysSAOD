```python
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt2

x  = [10000, 20000, 30000, 40000, 50000]
y1 = [224.17, 988.43, 2272.39, 4155.18, 6479.58]
y2 = [41.51, 169.61, 398.94, 726.87, 1108.66]
plt.plot(x, y1, label="Bouble sort")
plt.plot(x, y2, label="Insert sort")

plt.xlabel("Size")
plt.ylabel("Time")
plt.title("BigO Sorts")
plt.grid(1)
plt.legend()
plt.show()

x2 = [10000, 20000, 30000, 40000, 50000]
y3 = [6.07, 12.72, 19.46, 30.82, 36.94]
y4 = [0.47, 1.62, 2.7, 4.02, 5.51]
y5 = [1.31, 2.4, 4.25, 6.78, 8.15]
plt2.plot(x2, y3, label="Merge sort")
plt2.plot(x2, y4, label="Quick sort")
plt2.plot(x2, y5, label="Shell sort")
plt2.xlabel("Size")
plt2.ylabel("Time")
plt2.title("BigO sort")
plt2.grid(1)
plt2.legend()
plt2.show()
```
